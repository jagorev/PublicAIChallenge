## inferenza con il modello addestrato

1. Caricare il pickle che contiene il modello e il LabelEncoder usato.
2. Estrarre il modello e l’encoder.
3. Passare al modello l’input da predire, in formato DataFrame.
4. Decodificare l’output del modello con il LabelEncoder.


```python
import pandas as pd
# a sample taken from the test set.
record = pd.read_csv("sample_input.csv")
```

```python
import pickle
with open("model.pkl", "rb") as f:
    saved = pickle.load(f)

model = saved["model"]
le = saved["label_encoder"]
```

```python
pred = model.predict(record)
decoded = le.inverse_transform(pred)
print(decoded)
```
